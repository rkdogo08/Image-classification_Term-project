```
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```
#path="/content/image_0002.jpg"
#path="/content/gdrive/My Drive/Data_Pattern/data/test/Bedroom/image_0007.jpg"
path="/content/gdrive/My Drive/Data_Pattern/data/test/"
train_class=os.listdir(path)


file=path+train_class[0]+"/image_0007.jpg"

img=cv2.imread(file)

```

```

gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
 
sift = cv2.xfeatures2d.SIFT_create()
kp = sift.detect(gray,None)
 
cv2.drawKeypoints(gray,kp,img,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img)
```
